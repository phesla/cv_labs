```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./pseudo_colour/img/data.png')
```

```python
plt.imshow(img)
```

```python
plt.imshow(img[:, :, 1])
```

```python
# make red ang green images 
red_img = np.zeros(img.shape)
red_img[:, :, :] = [255, 0, 0]
green_img = np.zeros(img.shape)
green_img[:, :, :] = [0, 255, 0]
```

```python
plt.imshow(red_img)
```

```python
plt.imshow(green_img)
```

```python
# make red text image
red = (img >= 200).astype(np.uint16) * 255
red = cv2.bitwise_or(red_img.astype(np.uint16), red)
```

```python
# make green text image
green = (img >= 200).astype(np.uint16) * 255
green = cv2.bitwise_or(green_img.astype(np.uint16), green)
```

```python
plt.imshow(red)
```

```python
plt.imshow(green)
```

```python
# rotate green text image
green_rotated = np.rot90(green)
```

```python
plt.imshow(green_rotated)
```

```python
# resize and apply bitwise and operation to images
result_img = cv2.resize(green_rotated, (1000, 1000)) & cv2.resize(red, (1000, 1000))
```

```python
plt.imshow(result_img)
```
